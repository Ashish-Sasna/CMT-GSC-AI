```python
amangal_df = util.pd.read_csv(data_path + '/Amangal dataset/proc_elem_stream_sediments_gcs_ngdr_56L05.csv')
```

```python
def regr_model(obj, X, y, grid, cv, scoring):

    grid_search = util.GridSearchCV(obj, 
                                    grid, 
                                    cv=cv, 
                                    scoring=scoring, 
                                    verbose=5, 
                                    n_jobs=-1)
    
    grid_search.fit(X, y)

    return grid_search
```

```python
def prediction_comparison(data_df, predictors_df, target_elem, dimension, cv, scoring, rf_param_grid, xgb_param_grid):

    perf_dict = {
        'Percentage' : [],
        'Predictors' : [],
        'RF_y_pred' : [],
        'XGB_y_pred' : []
    }
    
    for i in range(5, 101, 5):

        # Load predictors
        pred_list = util.select_predictors(predictors_df, i, target_elem, dimension)

        # Load pre-processed data
        elem = util.elem_list(data_df)
        elem_df = data_df[elem]

        # Split data into train and test
        target = data_df[target_elem]
        train_df = elem_df[pred_list]

        X_train, X_test, y_train, y_test = util.train_test_split(train_df, 
                                                                 target, 
                                                                 test_size=0.15, 
                                                                 random_state=42,
                                                                 shuffle=True)

        # Normalize the data
        rs = util.RobustScaler()
        norm_data = rs.fit_transform(X_train)
        X_train_norm = util.pd.DataFrame(norm_data, 
                                         index=X_train.index, 
                                         columns=X_train.columns)

        # Random Forest
        rf = util.RandomForestRegressor()
        rf_model = regr_model(rf, X_train_norm, y_train, rf_param_grid, cv, scoring)
        y_pred_rf = util.pred_val(X_test, rs, rf_model)

        # XGBoost
        xgb = util.xgboost.XGBRFRegressor()
        xgb_model = regr_model(xgb, X_train_norm, y_train, xgb_param_grid, cv, scoring)
        y_pred_xgb = util.pred_val(X_test, rs, xgb_model)

        #Store the results based on percentage of predictors selected
        perf_dict['Percentage'].append(i)
        perf_dict['Predictors'].append(pred_list)
        perf_dict['RF_y_pred'].append(y_pred_rf)
        perf_dict['XGB_y_pred'].append(y_pred_xgb)

    # Return the dataframe of results
    perf_df = util.pd.DataFrame(perf_dict)
    index = y_test.index.tolist()
    y_test = y_test.tolist()

    return [perf_df, y_test, index]
```

```python
rf_param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_features': [1, 2, 3, 4, 'sqrt'],   # Number of features to consider at every split
    'max_depth': [10, 20, 30],    # Maximum number of levels in each tree
    'min_samples_split': [2, 5, 10],            # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4]              # Minimum number of samples required at each leaf node
}

xgb_param_grid = {
    'n_estimators': [100, 200, 300],    
    'learning_rate': [0.01, 0.05, 0.1],  
    'max_depth': [3, 5, 7],          
    'colsample_bytree': [0.5, 0.7, 1.0], 
    'subsample': [0.5, 0.7, 1.0],        
    'gamma': [0.1, 0.2],             
    'reg_alpha': [0.1, 0.5, 1],       
    'reg_lambda': [0.1, 0.5, 1],           
    'objective': ['reg:squarederror']
}

ce_2D_pred_df, y_test, index = prediction_comparison(amangal_df, pca_2d_ce, 'ce', '2D', 5, 'neg_mean_absolute_error', rf_param_grid, xgb_param_grid)
```

```python
ce_2D_pred_df['rf_error'] = ce_2D_pred_df.apply(lambda row: [(y_pred - y_test) for y_pred, y_test in zip(row['RF_y_pred'], 
                                                                                                         y_test)], 
                                                axis=1)
ce_2D_pred_df['xgb_error'] = ce_2D_pred_df.apply(lambda row: [(y_pred - y_test) for y_pred, y_test in zip(row['XGB_y_pred'], 
                                                                                                         y_test)], 
                                                axis=1)
```

```python
ce_2D_pred_df.head()
```

```python
def calculate_mae(row, y, model):

    y_test = y
    y_pred = row[f'{model}_y_pred']
    error = util.np.round(util.mean_absolute_error(y_test, y_pred))

    return error

ce_2D_pred_df['rf_mae'] = ce_2D_pred_df.apply(lambda row: calculate_mae(row, y_test, 'RF'), axis=1)
ce_2D_pred_df['xgb_mae'] = ce_2D_pred_df.apply(lambda row: calculate_mae(row, y_test, 'XGB'), axis=1)
```

```python
ce_2D_pred_df['Feature labels'] = ce_2D_pred_df.apply(lambda row:f"{row['Percentage']}%[{','.join(row['Predictors'])}]", axis=1)
```

### Error plots based on predictors selected on proximity from target

```python
error_data_rf = {
    'Test_Index' : [],
    'Percentage' : [],
    'Error' : [], 
    'Feature labels' : []
}

error_data_xgb = {
    'Test_Index' : [],
    'Percentage' : [],
    'Error' : [], 
    'Feature labels' : []
}
```

```python
for idx, row in ce_2D_pred_df.iterrows():
    # percentage_feature = [str(row['Percentage']) + '%', row['Predictors']]
    percentage = row['Percentage']
    for idx, error in enumerate(row['rf_error']):
        error_data_rf['Test_Index'].append(index[idx])
        error_data_rf['Percentage'].append(percentage)
        error_data_rf['Error'].append(error)
        error_data_rf['Feature labels'].append(row['Feature labels'])
    for idx, error in enumerate(row['xgb_error']):
        error_data_xgb['Test_Index'].append(index[idx])
        error_data_xgb['Percentage'].append(percentage)
        error_data_xgb['Error'].append(error)
        error_data_xgb['Feature labels'].append(row['Feature labels'])
```

```python
error_data_rf_df = util.pd.DataFrame(error_data_rf)
error_data_xgb_df = util.pd.DataFrame(error_data_xgb)
```

```python
error_data_rf_df1 = error_data_rf_df[(error_data_rf_df['Percentage'] >= 5) & (error_data_rf_df['Percentage'] <= 15)]
error_data_rf_df2 = error_data_rf_df[(error_data_rf_df['Percentage'] >= 35) & (error_data_rf_df['Percentage'] <= 50)]
error_data_rf_df3 = error_data_rf_df[(error_data_rf_df['Percentage'] == 75)]

error_data_rf_df_new = util.pd.concat([error_data_rf_df1, error_data_rf_df2, error_data_rf_df3], axis=0, ignore_index=True)
error_data_rf_df_new
```
